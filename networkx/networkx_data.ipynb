# Networkx data

## 0. Import

```python
import networkx as nx
```

## 1. Data

```python
G = nx.Graph()
G.add_edge('A', 'B')
G.add_edge('A', 'D')
G.add_edge('B', 'D')
G.add_edge('A', 'C')
G.add_edge('C', 'D')
G.add_edge('C', 'E')
G.add_edge('C', 'F')
G.add_edge('C', 'G')
nx.draw_networkx(G)
```

```python
G.nodes()
```

```python
G.edges()
```

```python
for adjacencies in G.adjacency():
    print(adjacencies)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 12))

# position
pos = nx.spring_layout(G, k=0.3)

# node size
node_size = []
for node, adjacencies in enumerate(G.adjacency()):
    node_size.append(len(adjacencies[1]) * 200)
nx.draw_networkx_nodes(G, pos, node_color="w", edgecolors='black', alpha=0.6, node_size=node_size)

# draw labels
nx.draw_networkx_labels(G, pos)

# edge width
edge_width = []
for node, adjacencies in enumerate(G.adjacency()):
    edge_width.append(len(adjacencies[1]))
nx.draw_networkx_edges(G, pos, alpha=0.4, edge_color="c", width=edge_width)
```
