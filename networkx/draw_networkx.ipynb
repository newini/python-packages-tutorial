## 0. Import

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## 1. Create random network data

- n : int, Number of nodes
- radius: float, Distance threshold value
- dim : int, optional, Dimension of graph
- pos : dict, optional, A dictionary keyed by node with node positions as values
- p: float, optional, Which Minkowski distance metric to use. p has to meet the condition 1 <= p <= infinity
- seed: integer, random_state, or None (default), Indicator of random number generation state. See Randomness.

```python
G = nx.random_geometric_graph(n=100, radius=0.125, seed=42)
```

## 2. Default draw_networkx

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 12))
nx.draw_networkx(G)
```

## 3. Make beutiful

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 12))

# position
pos = nx.spring_layout(G, k=0.3)

# node size
node_size = []
for node, adjacencies in enumerate(G.adjacency()):
    node_size.append(len(adjacencies[1]) * 200)
nx.draw_networkx_nodes(G, pos, node_color="w", edgecolors='black', alpha=0.6, node_size=node_size)

# draw labels
nx.draw_networkx_labels(G, pos)

# edge width
edge_width = []
for node, adjacencies in enumerate(G.adjacency()):
    edge_width.append(len(adjacencies[1]))
nx.draw_networkx_edges(G, pos, alpha=0.4, edge_color="c", width=edge_width)
```
