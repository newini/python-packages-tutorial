# Link analysis

- Pagerank

https://networkx.org/documentation/stable/reference/algorithms/link_analysis.html

## 0. Import

```python
import networkx as nx
```

## 1. Data

```python
G = nx.Graph()
G.add_edge('A', 'B')
G.add_edge('A', 'D')
G.add_edge('B', 'D')
G.add_edge('A', 'C')
G.add_edge('C', 'D')
G.add_edge('C', 'E')
G.add_edge('C', 'F')
G.add_edge('C', 'G')
nx.draw_networkx(G)
```

## Pagerank

```python
pr = nx.pagerank(G)
pr
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))

# position
pos = nx.spring_layout(G, k=0.3)

# node size
node_size = [5000*v for v in pr.values()]
nx.draw_networkx_nodes(G, pos, node_color="w", edgecolors='black', alpha=0.6, node_size=node_size)

# draw labels
nx.draw_networkx_labels(G, pos)

# draw edges
nx.draw_networkx_edges(G, pos, alpha=0.4, edge_color="c")
```

### use personalization

```python
personalization = {'A': 0, 'B': 0, 'D': 0, 'C': 1, 'E': 0, 'F': 0, 'G': 0}
pr = nx.pagerank(G, personalization=personalization)
pr
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))

# position
pos = nx.spring_layout(G, k=0.3)

# node size
node_size = [5000*v for v in pr.values()]
nx.draw_networkx_nodes(G, pos, node_color="w", edgecolors='black', alpha=0.6, node_size=node_size)

# draw labels
nx.draw_networkx_labels(G, pos)

# draw edges
nx.draw_networkx_edges(G, pos, alpha=0.4, edge_color="c")
```
